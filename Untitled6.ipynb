```python
import sqlalchemy as sa
import os

from zipline.finance import constants

from pluto.finance.commission import schema

from protos import slippage_pb2

def add_directory(name, root=None):
    if root:
        pth = os.paths.join(root, name)
    else:
        pth = name
    os.mkdir(pth)
    return pth

def add_file(name, root=None):
    if root:
        pth = os.paths.join(root, name)
    else:
        pth = name
    with open('pth', 'a'):
        pass
    return pth

pth = add_directory('setup', 'resources')
slippage_dir = add_directory('slippage', pth)
commission_dir = add_directory('commission', pth)

with open(os.paths.join(slippage_dir, 'volatility_volume_share'), 'a') as f:
    vvs = slippage_pb2.VolatilityVolumeShare(
        volume_limit=0.05,
        eta=constants.ROOT_SYMBOL_TO_ETA)
    f.write(vvs.SerializeToString())

models = add_file('models', commission_dir)

engine = sa.create_engin('sqlite:///{}'.format(models))
schema.metadata.create_all(engine)

def add_commission_model(broker, model, exchange, country_code, asset_type, parameters):
    models = schema.commission_metas
    models.insert().values(
        broker=broker, 
        model=model, 
        exchange=exchange, 
        country_code=country_code,
        asset_type=asset_type)
    
    table = schema[model]
    table.insert().values(
        **parameters
    )


data = {
    'test': {
        'country_code': {
        
    }}
}    
    
add_commission_model('test', '', 'NYSE')
```
